```python
raw_train_data=pd.read_csv("../input/titanic/train.csv")
raw_test_data=pd.read_csv("../input/titanic/test.csv")
```

```python
raw_train_data['Age']=raw_train_data['Age'].fillna(value=np.mean(raw_train_data['Age']))
raw_test_data['Age']=raw_test_data['Age'].fillna(value=np.mean(raw_test_data['Age']))

raw_train_data['Fare']=raw_train_data['Fare'].fillna(value=np.mean(raw_train_data['Fare']))
raw_test_data['Fare']=raw_test_data['Fare'].fillna(value=np.mean(raw_test_data['Fare']))
```

```python
raw_train_data['Cabin']=raw_train_data['Cabin'].fillna(0)
raw_test_data['Cabin']=raw_test_data['Cabin'].fillna(0)


raw_train_data['Cabin']=raw_train_data['Cabin'].replace(to_replace=0,value='XXX')
raw_test_data['Cabin']=raw_test_data['Cabin'].replace(to_replace=0,value='XXX')
```

```python
raw_train_data['Sex']=raw_train_data['Sex'].replace(to_replace='female',value='1',regex=True)
raw_train_data['Sex']=raw_train_data['Sex'].replace(to_replace='male',value='0',regex=True)
raw_test_data['Sex']=raw_test_data['Sex'].replace(to_replace='female',value='1',regex=True)
raw_test_data['Sex']=raw_test_data['Sex'].replace(to_replace='male',value='0',regex=True)
```

```python
train_cabin=raw_train_data['Cabin'].unique()
test_cabin=raw_test_data['Cabin'].unique()
cabin_combined=np.append(train_cabin,test_cabin)
```

```python
cabin=np.unique(cabin_combined)
```

```python
cabin.size
```

```python
cabin_val=np.arange(187)
```

```python
test_data=raw_test_data
train_data=raw_train_data
```

```python
train_data=train_data.replace(to_replace=cabin,value=cabin_val)
test_data=test_data.replace(to_replace=cabin,value=cabin_val)
```

```python
train_data['Embarked']=train_data['Embarked'].fillna(value='X')
emb=train_data['Embarked'].unique()
emb_val=np.arange(4)
train_data=train_data.replace(to_replace=emb,value=emb_val)
test_data=test_data.replace(to_replace=emb,value=emb_val)
```

```python
train_data=train_data.sort_values(by=['PassengerId'])
test_data=test_data.sort_values(by=['PassengerId'])
```

```python
X_train=train_data[['Pclass','Sex','Age','SibSp','Parch','Cabin','Fare','Embarked']]
X_test=test_data[['Pclass','Sex','Age','SibSp','Parch','Cabin','Fare','Embarked']]
y_train=train_data['Survived']
```

```python
from sklearn.preprocessing import StandardScaler
s1=StandardScaler()
s2=StandardScaler()
X_train=s1.fit(X_train).transform(X_train)
X_test=s2.fit(X_test).transform(X_test)
```
